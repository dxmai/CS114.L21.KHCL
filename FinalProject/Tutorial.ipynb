```python
import scrapy

class NewsSpider(scrapy.Spider):
    name = "news" #Đây là cái tên để 1 lát nữa muốn chạy chương trình thì gọi cái tên này trong terminal nà

    def start_requests(self):
      #urls là nơi tổng hợp các link. Ví dụ trang báo chỉ chứa 1 loại tin thì để 1 cái, nếu có nhiều loại tin
      #như là sport, entertainment,.... thì cho luôn các link trang đó vào
      #Ví dụ chỉ có 1 link thì nó sẽ như thế này nè
      '''
       urls = [
            'http://dailyheadlines.net/'
        ]
      '''
      #Còn nhiều link thì nó sẽ giống cái code dưới đây nè, mình vào từng mục rồi copy cái link của mục đó vào nhaaa
        urls = [
            'https://www.activistpost.com/',
            'https://www.activistpost.com/category/economy', 
            'https://www.activistpost.com/category/liberty', 
            'https://www.activistpost.com/category/war',
            'https://www.activistpost.com/category/activism',
            'https://www.activistpost.com/category/health',
            'https://www.activistpost.com/category/technology',
            'https://www.activistpost.com/category/environment'
        ]
        #Giờ thì với mỗi cái url mình lấy được, mình sẽ dùng cái link đó (chỗ url=url) á mọi người,
        #gọi hàm parse (chỗ callback á), nhớ thêm self phía trước vì mấy cái hàm này mình đều để trong 
        # 1 lớp (class là NewsSpider á, chỗ đầu tiên mình vào mình khai báo class)
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)
```

cái dòng đó hiểu là ở class **next**, tag **a** (tag là cái <..> </...> á), lấy cái **attr**ibute là **href**, mấy cái **attr**ibute nó thường có màu cam cam trong cái tag á mọi người)

```python
#Tui chính là tổng hợp những gì bạn Mai nói dài dòng nãy giờ á =..= copy tui rồi chạy là được à ;;-;;
import scrapy

class NewsSpider(scrapy.Spider):
    name = "news"

    def start_requests(self):
        urls = [
            'https://www.activistpost.com/'
        ]
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        getLinkInDiv = response.css('h3.content-list-title')
        allLinks = getLinkInDiv.css('a::attr(href)').getall()

        for each_link in allLinks:
            yield scrapy.Request(each_link, callback=self.parse_href)

        next_page = response.css('a.next::attr(href)').get()
        if next_page is not None:
            next_page = response.urljoin(next_page)
            yield scrapy.Request(next_page, callback=self.parse)


    def parse_href(self, response):
        title = response.css('.entry-title::text').get()
        title = title.replace('\n', '')
        title = title.replace('\t', '')
        author = response.css('.entry-content a::text').get()
        link = response.url
        date = response.css('.entry-meta span::text').get()
        #Nếu mà có lọc lại lấy từ năm bao nhiêu thì thêm vài câu lệnh ở đây để lọc ra nè ;;-;;
        paragraph = response.css('.entry-content p::text').getall()
        text = ''
        count = 0
        for sentence in paragraph:
            if count == 4:
                break
            if count == 0:
                count += 1
                continue
            sentencs = sentence.replace('\xa0', '')
            text = text + sentence
            count += 1
        yield {
            'author': author,
            'source': link,
            'date': date,
            'title': title,
            'text': text,
            'is_fake': 1
        }
```

```python
import scrapy
offset, times = 0, 0
valid = True
res = ''
dictionary = {
    'national/': [1003, 25]
}


class NewsSpider(scrapy.Spider):
    name = "news"

    def start_requests(self):
        global times, res, offset, valid
        urls = [
            'https://www.npr.org/sections/national/'
        ]

        for url in urls:
            for key in dictionary.keys():
                if url.endswith(key):
                    res = key
                    break
            times = 0
            offset = 0
            valid = True
            yield scrapy.Request(url, callback=self.parse)

    def parse(self, response):
        global offset, times
        getLinkInDiv = response.css('h2.title')
        allLinks = getLinkInDiv.css('a::attr(href)').getall()

        for each_link in allLinks:
            yield scrapy.Request(each_link, callback=self.parse_href)

        if valid:
            next_page = 'https://www.npr.org/get/'
            temp = dictionary[res][0]
            next_page = next_page + str(temp)
            next_page = next_page + '/render/partial/next?start='
            if times == 0:
                offset += dictionary[res][1]
                times += 1
            else:
                offset += 24
            next_page = next_page + str(offset)
            next_page = response.urljoin(next_page)
            yield scrapy.Request(next_page, callback=self.parse)

    def parse_href(self, response):
        global valid
        yearsList = ['2021', '2020', '2019', '2018']
        available = False
        date = response.css('.date::text').get()
        for year in yearsList:
            if year in date:
                available = True
                break
        if available:
            title = response.css('.storytitle h1::text').get()
            paragraph = response.css('.storylocation p::text').getall()
            text = ''
            count = 0
            for sentence in paragraph:
                if count == 0 or count == 1 or count == 2:
                    count += 1
                    continue
                if count == 6:
                    break
                sentence = sentence.replace('\xa0', '')
                sentence = sentence.replace('\n', '')
                sentence = sentence.replace('\r', '')
                sentence = sentence.replace('\t', '')
                text = text + sentence
                count += 1
            yield {
                'date': date,
                'title': title,
                'text': text,
                'is_fake': 0
            }
        else:
            valid = False
```
