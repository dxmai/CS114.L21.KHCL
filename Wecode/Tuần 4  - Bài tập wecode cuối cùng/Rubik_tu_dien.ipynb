RUBIK TỨ DIỆN

```python
from sys import stdin

color = stdin.readline().split()
step = stdin.readline().split()
end = stdin.readline()

pyraminx = []
for i in range(4):
    temp = [color[i] for j in range(9)]
    pyraminx.append(temp)

leftSide = [4, 6, 5, 1]
upSide = [0, 1, 2, 3]
rightSide = [8, 3, 7, 6]

for i in range((len(step) - 1), -1, -1):
    if step[i] == "U'":
        temp = pyraminx[0][0]
        pyraminx[0][0] = pyraminx[2][4]
        pyraminx[2][4] = pyraminx[3][8]
        pyraminx[3][8] = temp
    elif step[i] == "u'":
        temp = []
        for j in range(4):
            temp.append(pyraminx[0][upSide[j]])
        for j in range(4):
            pyraminx[0][upSide[j]] = pyraminx[2][leftSide[j]]
            pyraminx[2][leftSide[j]] = pyraminx[3][rightSide[j]]
            pyraminx[3][rightSide[j]] = temp[j]
    elif step[i] == "U":
        temp = pyraminx[0][0]
        pyraminx[0][0] = pyraminx[3][8]
        pyraminx[3][8] = pyraminx[2][4]
        pyraminx[2][4] = temp
    elif step[i] == "u":
        temp = []
        for j in range(4):
            temp.append(pyraminx[0][upSide[j]])
        for j in range(4):
            pyraminx[0][upSide[j]] = pyraminx[3][rightSide[j]]
            pyraminx[3][rightSide[j]] = pyraminx[2][leftSide[j]]
            pyraminx[2][leftSide[j]] = temp[j]

    elif step[i] == "B'":
        temp = pyraminx[1][0]
        pyraminx[1][0] = pyraminx[3][4]
        pyraminx[3][4] = pyraminx[2][8]
        pyraminx[2][8] = temp

    elif step[i] == "b'":
        temp = []
        for j in range(4):
            temp.append(pyraminx[1][upSide[j]])
        for j in range(4):
            pyraminx[1][upSide[j]] = pyraminx[3][leftSide[j]]
            pyraminx[3][leftSide[j]] = pyraminx[2][rightSide[j]]
            pyraminx[2][rightSide[j]] = temp[j]

    elif step[i] == "B":
        temp = pyraminx[1][0]
        pyraminx[1][0] = pyraminx[2][8]
        pyraminx[2][8] = pyraminx[3][4]
        pyraminx[3][4] = temp

    elif step[i] == "b":
        temp = []
        for j in range(4):
            temp.append(pyraminx[1][upSide[j]])
        for j in range(4):
            pyraminx[1][upSide[j]] = pyraminx[2][rightSide[j]]
            pyraminx[2][rightSide[j]] = pyraminx[3][leftSide[j]]
            pyraminx[3][leftSide[j]] = temp[j]


    elif step[i] == "L'":
        temp = pyraminx[2][0]
        pyraminx[2][0] = pyraminx[0][4]
        pyraminx[0][4] = pyraminx[1][8]
        pyraminx[1][8] = temp

    elif step[i] == "l'":
        temp = []
        for j in range(4):
            temp.append(pyraminx[2][upSide[j]])
        for j in range(4):
            pyraminx[2][upSide[j]] = pyraminx[0][leftSide[j]]
            pyraminx[0][leftSide[j]] = pyraminx[1][rightSide[j]]
            pyraminx[1][rightSide[j]] = temp[j]

    elif step[i] == "L":
        temp = pyraminx[2][0]
        pyraminx[2][0] = pyraminx[1][8]
        pyraminx[1][8] = pyraminx[0][4]
        pyraminx[0][4] = temp

    elif step[i] == "l":
        temp = []
        for j in range(4):
            temp.append(pyraminx[2][upSide[j]])
        for j in range(4):
            pyraminx[2][upSide[j]] = pyraminx[1][rightSide[j]]
            pyraminx[1][rightSide[j]] = pyraminx[0][leftSide[j]]
            pyraminx[0][leftSide[j]] = temp[j]

    elif step[i] == "R'":
        temp = pyraminx[3][0]
        pyraminx[3][0] = pyraminx[1][4]
        pyraminx[1][4] = pyraminx[0][8]
        pyraminx[0][8] = temp

    elif step[i] == "r'":
        temp = []
        for j in range(4):
            temp.append(pyraminx[3][upSide[j]])
        for j in range(4):
            pyraminx[3][upSide[j]] = pyraminx[1][leftSide[j]]
            pyraminx[1][leftSide[j]] = pyraminx[0][rightSide[j]]
            pyraminx[0][rightSide[j]] = temp[j]
    elif step[i] == "R":
        temp = pyraminx[3][0]
        pyraminx[3][0] = pyraminx[0][8]
        pyraminx[0][8] = pyraminx[1][4]
        pyraminx[1][4] = temp
    else:
        temp = []
        for j in range(4):
            temp.append(pyraminx[3][upSide[j]])
        for j in range(4):
            pyraminx[3][upSide[j]] = pyraminx[0][rightSide[j]]
            pyraminx[0][rightSide[j]] = pyraminx[1][leftSide[j]]
            pyraminx[1][leftSide[j]] = temp[j]

for i in range(len(pyraminx)):
    print(*pyraminx[i], sep = ' ')
```
