```python
soTestcase = int(input())
count = 0
while (count < soTestcase):
    array1 = [0] * 26
    array2 = [0] * 26
    string1 = input()
    string2 = input()

    for i in string1:
        array1[ord(i) - ord('a')] += 1

    for j in string2:
        array2[ord(j) - ord('a')] += 1

    index = 0
    flag = False
    while(index < len(array1)):
        if array1[index] != 0 and array2[index] != 0:
            flag = True
            break
        index += 1

    if flag:
        print("YES")
    else:
        print("NO")

    count += 1
```

Ý tưởng giải bài toán: 2 chuỗi chỉ có thể biến đổi để giống nhau khi cả 2 chuỗi có cùng ít nhất là 1 ký tự => Đếm số lượng ký tự trong mỗi chuỗi, nếu có 1 ký tự cùng xuất hiện ở cả hai chuỗi thì chắc chắn có thể biến đổi được, nếu đi hết độ dài chuỗi vẫn không tìm được thì không thể biến đổi
