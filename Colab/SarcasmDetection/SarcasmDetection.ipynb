### Trang newsbiscuit.com (Anh)

```python
import scrapy


class NewsSpider(scrapy.Spider):
    name = "news"

    def start_requests(self):
        urls = [
            'http://www.newsbiscuit.com/category/arts-and-entertainment/',
            'http://www.newsbiscuit.com/category/business/',
            'http://www.newsbiscuit.com/category/celebrity/',
            'http://www.newsbiscuit.com/category/crime/',
            'http://www.newsbiscuit.com/category/education/',
            'http://www.newsbiscuit.com/category/environment/',
            'http://www.newsbiscuit.com/category/faith-2/',
            'http://www.newsbiscuit.com/category/features/',
            'http://www.newsbiscuit.com/category/from-the-archives/',
            'http://www.newsbiscuit.com/category/health/',
            'http://www.newsbiscuit.com/category/isle-of-wight-news/',
            'http://www.newsbiscuit.com/category/latest/',
            'http://www.newsbiscuit.com/category/lifestyle/',
            'http://www.newsbiscuit.com/category/politics/',
            'http://www.newsbiscuit.com/category/science-and-technology/',
            'http://www.newsbiscuit.com/category/sport/',
            'http://www.newsbiscuit.com/category/stuff/',
            'http://www.newsbiscuit.com/category/uk-news/',
            'http://www.newsbiscuit.com/category/world-news/'
        ]

        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        years_list = ['2021', '2020', '2019', '2018']
        flag = True
        for post in response.css('div.articlebox'):
            year = post.css('time::attr(datetime)').get()
            year = year[0:4]
            if year in years_list:
                yield {
                    'headline': post.css('.col-sm-8 a::text').get(),
                    'article_link': post.css('.col-sm-8 a::attr(href)').get(),
                    'is_sarcastic': 1,
                }
            else:
                flag = False

        if flag:
            next_page = response.css('.col-xs-6 a::attr(href)').get()
            if next_page is not None:
                next_page = response.urljoin(next_page)
                yield scrapy.Request(next_page, callback=self.parse)

```

##Trang thedailymash.com

```python
import scrapy


class NewsSpider(scrapy.Spider):
    name = "news"

    def start_requests(self):
        urls = [
            'https://www.thedailymash.co.uk/news',
            'https://www.thedailymash.co.uk/news/lifestyle',
            'https://www.thedailymash.co.uk/news/society',
            'https://www.thedailymash.co.uk/news/health',
            'https://www.thedailymash.co.uk/news/arts-entertainment',
            'https://www.thedailymash.co.uk/news/science-technology',
            'https://www.thedailymash.co.uk/news/business',
            'https://www.thedailymash.co.uk/news/relationships',
            'https://www.thedailymash.co.uk/politics',
            'https://www.thedailymash.co.uk/sport'
        ]

        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        name = 'div.md\:mr-8'
        for post in response.css(name):
            yield {
                'headline': post.css('.-mt-1 a::text').get(),
                'article_link': 'https://www.thedailymash.co.uk' + post.css('.-mt-1 a::attr(href)').get(),
                'is_sarcastic': 1
            }

        next_page = response.css('a.ml-3::attr(href)').get()
        if next_page is not None:
            next_page = response.urljoin(next_page)
            yield scrapy.Request(next_page, callback=self.parse)

```
