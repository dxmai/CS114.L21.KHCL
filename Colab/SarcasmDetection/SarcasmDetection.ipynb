```python
import scrapy

class NewsSpider(scrapy.Spider):
    name = "news"
    def start_requests(self):
        urls = [
            'https://www.thebeaverton.com/',
        ]
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        for post in response.css('div.medium-7'):
            yield {
                'title': post.css('.post-title a::text').get(),
                'link': post.css('.post-title a::attr(href)').get(),
                'label': 1
            }
        next_page = response.css('a.next::attr(href)').get()
        if next_page is not None:
            next_page = response.urljoin(next_page)
            yield scrapy.Request(next_page, callback=self.parse)
```

```python
import scrapy


class NewsSpider(scrapy.Spider):
    name = "news"

    def start_requests(self):
        urls = [
            'https://www.thedailymash.co.uk/news',
            'https://www.thedailymash.co.uk/news/lifestyle',
            'https://www.thedailymash.co.uk/news/society',
            'https://www.thedailymash.co.uk/news/health',
            'https://www.thedailymash.co.uk/news/arts-entertainment',
            'https://www.thedailymash.co.uk/news/science-technology',
            'https://www.thedailymash.co.uk/news/business',
            'https://www.thedailymash.co.uk/news/relationships',
            'https://www.thedailymash.co.uk/politics',
            'https://www.thedailymash.co.uk/sport'
        ]

        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        name = 'div.md\:mr-8'
        for post in response.css(name):
            yield {
                'headline': post.css('.-mt-1 a::text').get(),
                'article_link': 'https://www.thedailymash.co.uk' + post.css('.-mt-1 a::attr(href)').get(),
                'is_sarcastic': 1
            }

        next_page = response.css('a.ml-3::attr(href)').get()
        if next_page is not None:
            next_page = response.urljoin(next_page)
            yield scrapy.Request(next_page, callback=self.parse)

```
