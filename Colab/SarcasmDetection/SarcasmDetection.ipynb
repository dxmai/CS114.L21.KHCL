```python
import scrapy

class NewsSpider(scrapy.Spider):
    name = "news"
    def start_requests(self):
        urls = [
            'https://www.thebeaverton.com/',
        ]
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        for post in response.css('div.medium-7'):
            yield {
                'title': post.css('.post-title a::text').get(),
                'link': post.css('.post-title a::attr(href)').get(),
                'label': 1
            }
        next_page = response.css('a.next::attr(href)').get()
        if next_page is not None:
            next_page = response.urljoin(next_page)
            yield scrapy.Request(next_page, callback=self.parse)
```

Câu lệnh chạy trên terminal: 
scrapy crawl news -o thebeaverton.json

```python
import scrapy


class NewsSpider(scrapy.Spider):
    name = "news"

    def start_requests(self):
        urls = [
            'https://www.theonion.com/local',
            'https://www.theonion.com/breaking-news/news-in-brief',
            'https://www.theonion.com/entertainment/magazine',
            'https://www.theonion.com/politics/news-in-brief',
            'https://www.theonion.com/sports/basketball',
            'https://www.theonion.com/sports/football',
            'https://www.theonion.com/sports/baseball',
            'https://www.theonion.com/sports/hockey',
            'https://www.theonion.com/sports/golf',
            'https://www.theonion.com/sports/horse-racing'
        ]

        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse)

    def parse(self, response):
        for post in response.css('div.cw4lnv-5'):
            yield {
                'title': post.css('h2::text').get(),
                'link': post.css('a::attr(href)').get(),
                'label': 1
            }

        next_page = response.css('a.next-button::attr(href)').get()
        if next_page is not None:
            next_page = response.urljoin(next_page)
            yield scrapy.Request(next_page, callback=self.parse)
```
